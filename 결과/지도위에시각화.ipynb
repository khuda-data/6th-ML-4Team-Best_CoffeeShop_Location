# 반경 800m

```python
# csv파일로 데이터프레임 만들기
df_800 = pd.read_excel('/content/drive/MyDrive/쿠다/쿠다 6기/토이프로젝트/cluster_1_sorted.xlsx')
df_800.head()
```

```python
cluster_1_sorted = df_800[df_800['density'] == 0]
```

```python
cluster_1_sorted.head()
```

```python
from shapely.geometry import Point
import geopandas as gpd

# centroid 문자열을 Point 객체로 변환
cluster_1_sorted['centroid'] = cluster_1_sorted['centroid'].apply(lambda x: gpd.GeoSeries.from_wkt([x])[0])
from matplotlib.colors import LinearSegmentedColormap

cluster_1_sorted['x'] = cluster_1_sorted['centroid'].apply(lambda point: point.x)
cluster_1_sorted['y'] = cluster_1_sorted['centroid'].apply(lambda point: point.y)
```

```python
cluster_1_sorted.head()
```

```python
import folium

# 지도 생성 (예시: 첫 번째 좌표를 기준으로 중앙을 잡습니다)
map = folium.Map(location=[cluster_1_sorted['y'].mean(), cluster_1_sorted['x'].mean()], zoom_start=13)

# iterrows()를 사용해 각 행을 반복하며 원을 추가
for _, row in cluster_1_sorted.iterrows():
    folium.Circle(
        location=[row['y'], row['x']],  # 원의 중심 좌표
        radius=400,                     # 반경 (미터 단위)
        color='red',                    # 원의 테두리 색상
        fill=True,                      # 원을 채울지 여부
        fill_color='red',               # 원 채우기 색상
        fill_opacity=0.4                # 원의 투명도 (0.4 = 40% 투명도)
    ).add_to(map)

# 지도를 출력합니다.
map
```

# 비교

```python
import folium
from folium.plugins import FloatImage
import branca

# 지도 생성 (고가형의 평균 좌표를 기준으로 중앙을 잡습니다)
map = folium.Map(location=[(df_고가['y'].mean() + df_고가['y'].mean()) / 2,
                           (df_고가['x'].mean() + df_고가['x'].mean()) / 2], zoom_start=13)

# 고가형 위치 표시 (빨간색)
for index, row in df_고가.iterrows():
    index = index + 1
    folium.Circle(
        location=[row['y'], row['x']],  # 원의 중심 좌표
        radius=150,                     # 반경 (미터 단위)
        color='red',                    # 원의 테두리 색상
        fill=True,                      # 원을 채울지 여부
        fill_color='red',               # 원 채우기 색상
        fill_opacity=0.4                # 원의 투명도 (0.4 = 40% 투명도)
    ).add_to(map)

    folium.Marker(
        location=[row['y'], row['x']],
        icon=folium.DivIcon(html=f"<div style='font-size: 12pt; color: black'>{index}</div>")
    ).add_to(map)

# 저가형 위치 표시 (파란색)
for index, row in df_저가.iterrows():
    index = index + 1
    folium.Circle(
        location=[row['y'], row['x']],  # 원의 중심 좌표
        radius=150,                     # 반경 (미터 단위)
        color='blue',                   # 원의 테두리 색상
        fill=True,                      # 원을 채울지 여부
        fill_color='blue',              # 원 채우기 색상
        fill_opacity=0.4                # 원의 투명도 (0.4 = 40% 투명도)
    ).add_to(map)

    folium.Marker(
        location=[row['y'], row['x']],
        icon=folium.DivIcon(html=f"<div style='font-size: 12pt; color: black'>{index}</div>")
    ).add_to(map)

# 범례 추가
legend_html = """
     <div style="
     position: fixed;
     bottom: 50px; left: 50px; width: 150px; height: 90px;
     background-color: white;
     border:2px solid grey; z-index:9999; font-size:14px;
     ">
     &nbsp; <b>Legend</b> <br>
     &nbsp; <i style="color:red">●</i> 고가형 <br>
     &nbsp; <i style="color:blue">●</i> 저가형 <br>
     </div>
     """
map.get_root().html.add_child(folium.Element(legend_html))

# 지도를 출력합니다.
map
```
