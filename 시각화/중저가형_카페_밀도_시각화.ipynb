```python
import io
# 데이터 불러오기
df_midlow = pd.read_csv('중저가형(최종)_cleaned (1).csv', encoding='cp949')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 그래프 설정
plt.figure(figsize=(10, 8))
# 중저가형 매장의 밀도 분석
sns.kdeplot(x=df_midlow['위도'], y=df_midlow['경도'],
            cmap="Reds", shade=True, bw_adjust=0.5, alpha=0.5, label='Low Price')
# 그래프 설정
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Density Analysis of Store Locations by Price Category')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_midlow_coords = df_midlow[['위도', '경도']]

# 산점도 그리기
plt.figure(figsize=(10, 8))
plt.scatter(df_midlow_coords['위도'], df_midlow_coords['경도'], c='red', label='Low Price', alpha=0.5)

plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Store Locations by Price Category (Using X, Y Coordinates)')
plt.legend()
plt.grid(True)
plt.show()
```
