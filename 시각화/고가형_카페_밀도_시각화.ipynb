```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files
uploaded = files.upload()
print(uploaded.keys())
```

```python
import io
# 데이터 불러오기
df_high = pd.read_excel(io.BytesIO(uploaded['high.xlsx']))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 그래프 설정
plt.figure(figsize=(10, 8))
# 고가형 매장의 밀도 분석
sns.kdeplot(x=df_high['위도'], y=df_high['경도'],
            cmap="Reds", shade=True, bw_adjust=0.5, alpha=0.5, label='High Price')
# 그래프 설정
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Density Analysis of Store Locations by Price Category')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_high_coords = df_high[['위도', '경도']]

# 산점도 그리기
plt.figure(figsize=(10, 8))
plt.scatter(df_high_coords['위도'], df_high_coords['경도'], c='red', label='High Price', alpha=0.5)

plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Store Locations by Price Category (Using X, Y Coordinates)')
plt.legend()
plt.grid(True)
plt.show()
```
